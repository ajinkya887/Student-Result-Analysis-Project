```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_score.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
#change weekly study hours column
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10", "5-10")
df.head()
```

#gender distribution

```python
plt.figure(figsize = (5,5))
custom_palette = ['green', 'orange']
ax = sns.countplot(data = df, x = "Gender", hue="Gender",palette=custom_palette)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that number of females are more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between parent's score and student's score")
plt.show()
```

```python
#from the above chart we have concluded that parents have good impact on childrens 
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parent's marital status and student's score")
plt.show()
```

```python
#from the above chart we have concluded that parents have negligible good impact on childrens 
```

```python
groupA = df.loc[(df['EthnicGroup'] =="group A")].count()
groupB = df.loc[(df['EthnicGroup'] =="group B")].count()
groupC = df.loc[(df['EthnicGroup'] =="group C")].count()
groupD = df.loc[(df['EthnicGroup'] =="group D")].count()
groupE = df.loc[(df['EthnicGroup'] =="group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mList = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

plt.pie(mList, labels = l, autopct = "%1.2f%%")
plt.title("Distribution Of Ethnic Groups")
plt.show()
```
